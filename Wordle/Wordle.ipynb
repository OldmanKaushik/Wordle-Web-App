```python
#task (wordle)
import random
with open('WORDS.txt','r') as file:
    file = [i.strip() for i in file.readlines()]
#for i in file:
    #print(i)
#file handling complete
word_loc = random.randint(0,len(file)-1) #randomly picks a word location
word = file[word_loc] #gets the word
#print(word) #testing the function
#testing worked

def validation(word):
    if len(word) != 5:
        print("does not meet length")
        return False
    elif word.isalpha() == False:
        print("There are symbols, other than letters in your input")
        return False
    else:
        return True
    
#best way to do this is using a dictionary
word_dic = {}
word_dic2 = {}

for i in range(6):
    for new in range(len(word)):
        word_dic[word[new]] = 0
    for inc in range(len(word)):
        word_dic[word[inc]] += 1
        
    input_word = input("Please guess a 5-letter word: ")
    while validation(input_word) == False:
        input_word = input("Please guess a 5-letter word: ")
        
    for new in range(len(input_word)):
        word_dic2[input_word[new]] = 0
    for inc in range(len(input_word)):
        word_dic2[input_word[inc]] += 1
        
    for j in range(len(input_word)):
        print(input_word[j], end="\t")
    print()
    for k in range(len(input_word)):
        if input_word[k] == word[k]:
            print("O", end="\t")
        else:
            if input_word[k] in word:
                if word_dic2[input_word[k]]> word_dic[input_word[k]]:
                    print("X", end="\t")
                    word_dic2[input_word[k]] -=1
                elif word_dic2[input_word[k]] <= word_dic[input_word[k]]:
                    print(" ", end="\t")
            else:
                print("X", end="\t")
    print()
    if word == input_word:
        print("Congrats! You have obtained the answer.")
        break
    else:
        print(f"Number of attempts left: {len(word)-i}")
        print()
print(word)
        
            

    
        
    
```
